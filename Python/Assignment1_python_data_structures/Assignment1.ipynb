Task 3

```python


price_dict = {
    "Laptop": 55000,
    "Mouse": 500,
    "Keyboard": 1200,
    "Monitor": 15000,
    "Printer": 8000,
    "Headphones": 2000
}

print("Initial price dictionary:", price_dict)


price_dict["Webcam"] = 2500

# Update price of existing product
price_dict["Mouse"] = 600

# Remove a product safely
product_to_remove = "Printer"
if product_to_remove in price_dict:
    del price_dict[product_to_remove]
    print(product_to_remove, "removed successfully.")
else:
    print(product_to_remove, "does not exist.")

print("Updated price dictionary:", price_dict)

# 3. Print average price
total_price = sum(price_dict.values())
average_price = total_price / len(price_dict)

print("Average price of products:", average_price)


# Product with maximum and minimum price
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)

print("Highest priced product:", max_product, "-", price_dict[max_product])
print("Lowest priced product:", min_product, "-", price_dict[min_product])
```

Task 4

```python
# Products list
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphones", "Webcam"]

# Category list (same order)
categories = ["Electronics", "Accessories", "Accessories",
              "Electronics", "Accessories", "Electronics"]

# Price dictionary
price_dict = {
    "Laptop": 55000,
    "Mouse": 600,
    "Keyboard": 1200,
    "Monitor": 15000,
    "Headphones": 2000,
    "Webcam": 2500
}


catalog = []

for i in range(len(products)):
    product_name = products[i]
    price = price_dict[product_name]
    category = categories[i]
    catalog.append((product_name, price, category))

print("Catalog:", catalog)

#  Create category_to_products dictionary
category_to_products = {}

for item in catalog:
    product_name = item[0]
    category = item[2]

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(product_name)

print("Category to Products:", category_to_products)

# Print products from category with maximum products
max_category = max(category_to_products, key=lambda k: len(category_to_products[k]))

print("Category with maximum products:", max_category)
print("Products in that category:", category_to_products[max_category])
```
